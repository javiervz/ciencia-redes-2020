```python
G = nx.Graph()
cantidad = 0
cocinas_en_red = {cuisine:0 for cuisine in cuisines}

# Agregamos los ingredientes al 
for receta in recetas:
    for ingrediente in receta['ingredients']:
        if not G.has_node(ingrediente):
            G.add_node(ingrediente)
            cantidad = cantidad + 1
            G.node[ingrediente]['cocinas'] = list()
            G.node[ingrediente]['cocinas'].append(receta['cuisine'])
            cocinas_en_red[receta['cuisine']] = cocinas_en_red[receta['cuisine']] + 1
        elif receta['cuisine'] in G.node[ingrediente]['cocinas']:
            G.node[ingrediente]['cocinas'].append(receta['cuisine'])
            cocinas_en_red[receta['cuisine']] = cocinas_en_red[receta['cuisine']] + 1

max = 0
for receta in recetas:
    for i in range(len(receta['ingredients'])):
        receta['ingredients'][i] = receta['ingredients'][i].replace(' ', '_')
        
for receta in recetas:
    for i in range(len(receta['ingredients'])):
        for j in range(i+1, len(receta['ingredients'])):
            if not G.has_edge(receta['ingredients'][i], receta['ingredients'][j]):
                G.add_edge(receta['ingredients'][i], receta['ingredients'][j], weight=1)
            else:
                G[receta['ingredients'][i]][receta['ingredients'][j]]['weight'] = G[receta['ingredients'][i]][receta['ingredients'][j]]['weight'] + 1
            
            if max < G[receta['ingredients'][i]][receta['ingredients'][j]]['weight']:
                max = G[receta['ingredients'][i]][receta['ingredients'][j]]['weight']


# Normalizamos los pesos
for edge in G.edges(data=True):
    edge[2]['weight'] = edge[2]['weight']/max

G2 = G
```

## Kmeans
Primero, para determinar el número de clusters usaremos el método del codo.


```python
from sklearn.cluster import KMeans

def calcula_codo(points, kmax):
  sse = []
  for k in range(1, kmax+1):
    kmeans = KMeans(n_clusters = k).fit(points)
    centroids = kmeans.cluster_centers_
    pred_clusters = kmeans.predict(points)
    curr_sse = 0
    
    # Se calcula el cuadrado de la distancia de cada punto al centro de su cluster
    for i in range(len(points)):
      curr_center = centroids[pred_clusters[i]]
      curr_sse += (points[i, 0] - curr_center[0]) ** 2 + (points[i, 1] - curr_center[1]) ** 2
      
    sse.append(curr_sse)
  return sse
```

```python
def calcula_silueta(puntos, kMax):
    from sklearn.metrics import silhouette_score

    sil = []

    # Se van agregando los puntajes de silueta a una lista que luego se retorna
    for k in range(2, kMax+1):
        kmeans = KMeans(n_clusters = k).fit(puntos)
        labels = kmeans.labels_
        sil.append(silhouette_score(puntos, labels, metric = 'euclidean'))
    return sil
```

```python

plt.plot(calcula_codo(node_embeddings_2d, 30))
plt.show()
```
