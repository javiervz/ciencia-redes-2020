Cuando estudiamos los sesgos exitentes dentro de las redes nos encontramos con la múltiples paradojas, como lo es la paradoja de la amistad, que estipula que, a un nivel perceptivo, todos los amigos de un individuo tienen cada uno más amigos que el individuo en sí. Una paradoja similar a esta en el espejismo de la mayoría en donde, por ejemplo, si un individuo ve un grupo de personas bailando en una fiesta, pensará que la mayoría estaba bailando aunque este grupo fuera solo el 20% de personas en la fiesta.

Definiéndolo en termino de redes, marcando ciertos nodos como activos, existe una probabilidad de que exista un espejismo de la mayoría en base a los nodos que tengan en su vecindad a este nodo activo. Esto produce que la percepción del nodo que observa al activo "piensa" que esta característica, que vuelve al nodo en un nodo activo, sea un común de la mayoría de la red a pesar de que no sea así.

Partimos creando los nodos

```python
import networkx as nx
import numpy as np
```

```python
v = 500
e = 5000

Gog = nx.erdos_renyi_graph(v, 2*np.log(v)/v) ## <-- por qué esta relación?

nx.is_connected(Gog)
```

```python
B = nx.barabasi_albert_graph(v, 100)
nx.is_connected(B)
```

Definimos los rangos de asertividad y un threshold de 0.005

```python
target_r = [-0.2,-0.1, 0, 0.1, 0.2]
threshold = 0.005
```

Modificamos la red para generar los rangos objetivos de asertividad

<span style="color:red"> 

- Explica un poco más tu metodología
    
</span>


```python
graphs = []
rs = []
for t in target_r:
    G = Gog.copy()
    for i in range(1000):
        r = nx.degree_pearson_correlation_coefficient(G)
        G0 = G.copy()
        nx.connected_double_edge_swap(G0)
        r0 = nx.degree_pearson_correlation_coefficient(G0)
        #check both sides
        if abs(r0 - t) < abs(r - t):
            G = G0.copy()
            r = r0
            #print(r)
        if abs(r - t) < threshold:
            break
    graphs.append(G.copy())
    rs.append(r)
    print(r)
```

Tomamos distintos porcentajes de nodos activos.
En este caso usamos un 1%, un 5% y un 10% de la red de manera aleatoria.

```python
from random import sample
active_proportions = [v*0.01, v*0.02, v*0.1]
```

```python
As = []
GAs = []
for active in active_proportions:
    random_nodes = sample(list(Gog.nodes()), int(active))
    print("-"*10)
    print("Using {} active nodes".format(int(active)))
    As = []
    for i, G_i in enumerate(graphs):
        print("Checking graph with r =", rs[i])
        A = 0
        for node in list(G_i.nodes()):
            adj = list(G_i.adj[node])
            inter = [value for value in adj if value in random_nodes]
            A_i = len(inter)/len(adj)
            A += A_i
        A = A/len(list(G_i.nodes()))
        As.append(A)
        print("The probability of mayority =", A)
    GAs.append(As)
```

Graficamos los resultados
### Probabilidad de Mayoría vs Espejismo

```python
import matplotlib.pyplot as plt
for i, y in enumerate(GAs):
    X = rs

    plt.plot(X, y)
    plt.title("{} Nodos activos".format(int(active_proportions[i])))
    plt.show()
```
