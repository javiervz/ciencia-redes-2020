### 1.b Describiendo los datos

Dado que se debe replicar parte de la metodología presentada en la investigación antes resumida, se utilizarán como fuente de datos seis archivos de texto, cada uno de los cuales contiene la Declaración Universal de Derechos Humanos en seis idiomas distintos. Cabe destacar que este documento fue proclamado por las Naciones Unidas, por lo que ha sido redactado y/o traducido en (al menos) los siguientes idiomas: árabe, chino, inglés, francés, ruso y español. Son justamente estos los idiomas de los documentos que se utilizarán en el presente trabajo.

A continuación, se pueden observar cuatro funciones:
- La primera (`get_data_archive`): lee el archivo y lo almacena en un arreglo en el que cada posición contiene una línea de texto del archivo. No se considera la cabecera del archivo, ya que no es relevante para el experimento. Luego cada línea se separa en un arreglo en el cual cada posición tiene una palabra.
- La segunda (`tokenize`): simplemente separa una cadena de texto que, posiblemente, tendrá más de una palabra (separadas por espacios) y la transforma en un arreglo en el cual cada posición contiene una palabra.
- La tercera (`clean_sentence`): a la oración recibida le elimina caracteres extraños, generalmente usados para puntuación. Transforma, si es posible, las letras en minúscula para estandarizar el texto.
- La cuarta (`clean`): recibe el texto completo en un idioma para limpiarlo, para ello se apoya de las dos funciones anteriores.

```python
import re
```

```python
def get_data_archive(langauge_filename):
    file = open('datos/' + langauge_filename, 'r', encoding = 'utf8')
    lines = file.read().split('\n')[6:]
    lines = [line.strip() for line in lines]
    lines = [line for line in lines if len(line) > 0]
    file.close()
    return lines

def tokenize(s):
    return s.split(' ')

def clean_sentence(sentence, language):
    strCleaner = '·×—‘’“”…、。《》『』【】！（），：；？一``!"#$%&\¿()*+,-./:;<=>?@[\\]_{|}'
    table = str.maketrans(strCleaner, ' ' * len(strCleaner))

    sentence = sentence.lower().translate(table)

    if(language == 'Chino'):
        sentence = re.sub(r'[a-z|A-Z|0-9]', ' ', sentence)

    return sentence

def clean(lines, language):
    lines = [tokenize(clean_sentence(sentence, language)) for sentence in lines]
    lines = [[word for word in sentence if not word.isdigit() and len(word) > 0] for sentence in lines]

    return lines
```

```python
language_filenames = {
    'Árabe': 'udhr_arb.txt',
    'Chino': 'udhr_cmn_hans.txt',
    'Inglés': 'udhr_eng.txt',
    'Francés': 'udhr_fra.txt',
    'Ruso': 'udhr_rus.txt',
    'Español': 'udhr_spa.txt'
}

language_lines = dict()

for language, filename in language_filenames.items():
    lines = get_data_archive(filename)
    language_lines[language] = clean(lines, language)
```

Como se puede haber notado, la tokenización del texto hasta ahora solo ha sido en base a los espacios que hay entre palabras. Sin embargo, en el chino no necesariamente se usan espacios para separar palabras, de hecho solo se usan de forma recurrente para los finales de frases o párrafos (por ejemplo, luego de una puntuación). Esto quiere decir que un conjunto de caracteres sucesivos (sin espacios entre medio) pueden representar varias palabras, lo que genera un problema a la hora de comparar este lenguaje con otros computacionalmente. Para intentar paliar en parte este problema, se ha utilizado una librería llamada Jieba, la cual se encarga de segmentar los caracteres en chino, con el objeto de formar palabras con estos, y así poder identificar cada palabra por separado de una oración, en la que antes todos los caracteres estaban juntos y no se podían disitinguir.

```python
import jieba.posseg as pseg
import jieba
```

```python
def tokenize_cmn(lines):
    lines_tokenized = list()
    for sentence in lines:
        new_sentence = list()
        lines_tokenized.append(new_sentence)
        for word in sentence:
            words = pseg.cut(word)
            for w in words:
                new_sentence.append(w.word)
    return lines_tokenized
```

```python
language_lines['Chino'] = tokenize_cmn(language_lines['Chino'])
```

### 2.a Construyendo las redes para cada lengua

A continuación, se construyen seis redes dirigidas: una para cada idioma. Cada red está compuesta por los nodos que representan a las distintas palabras del texto. Estas palabras están conectadas por aristas dirigidas y con peso, una arista está dirigida desde una palabra hacia la palabra que le sigue inmediatamente. El peso está dado por la cantidad de veces que una cierta palabra precede (in-degree) o sucede (out-degree) a otra en las frases del texto.

```python
import networkx as nx
```

```python
def create_graph(lines, limit = None):
    G = nx.DiGraph()

    for idx, sentence in enumerate(lines):
        G.add_nodes_from(sentence)
        if(len(sentence) > 1):
            pairs = list(zip(sentence, sentence[1:]))
            for pair in pairs:
                if(G.has_edge(pair[0], pair[1])):
                    G[pair[0]][pair[1]]['weight'] += 1
                else:
                    G.add_edge(pair[0], pair[1], weight = 1)
        if(limit and idx == limit):
            break
    return G
```

```python
language_graphs = dict()

for language, lines in language_lines.items():
    language_graphs[language] = create_graph(lines)

language_graphs
```

A modo de ejemplo se representa visualmente una red para el idioma español que considera tan solo dos frases. En esta se puede ver claramente cómo los nodos de palabras se conectan unos con otros, incluso se logra leer algo coherente siguiendo la dirección de las flechas (aristas). Por lo demás existen algunas palabras mejor conectadas que otras, por ejemplo "de". Sin embargo, también hay aquellas que no se conectan con ninguna otra como, por ejemplo, "preámbulo", esto es importante de destacar porque más adelante se considerarán los resultados estadísticos de las redes fuertemente conectadas más grandes, dejando de lado estás subredes pequeñas y débilmente conectadas.

```python
import matplotlib.pyplot as plt
```

```python
def plot_graph(G):
    fig, ax = plt.subplots(dpi = 800)
    pos = nx.circular_layout(G)
    labels = {i: i for i in G.nodes}
    
    nx.draw_networkx_nodes(G, pos, node_size = 100, node_color = 'orange') 
    nx.draw_networkx_edges(G, pos, alpha = 1, width = 0.5, edge_color = 'k')
    nx.draw_networkx_labels(G, pos, labels, alpha = 0.95, font_size = 5, font_color = 'k')

    plt.axis('off')
    plt.show()
```

```python
example_graph = create_graph(language_lines['Español'], 2)
plot_graph(example_graph)
```

### 2.d Replicando algunos experimentos del *paper*

A partir de cada una de las seis redes se obtienen sus subredes más grandes (con mayor cantidad de nodos). Cada subred corresponde a la componente fuertemente conectada (conexa) más grande encontrada en la red base de los idiomas.

En la tabla de a continuación se presentan las siguientes medidas relevantes que describen estas subredes fuertemente conexas:
- Cantidad de nodos.
- Cantidad de aristas.
- Peso total de las aristas: obtenido de la suma de los pesos de todas las aristas.
- Peso promedio de las aristas.
- Promedio de largo de los caminos más cortos.
- Promedio *in-degree*: promedio de grados de los nodos de una red considerando sus aristas "entrantes". Recordar que una arista "entrante" representa la conexión de una palabra con su predecesora.
- Mayor *in-degree*: el grado mayor presente en la red del nodo considerando sus aristas "entrantes".
- Promedio *out-degree*: promedio de grados de los nodos de una red considerando sus aristas "salientes". Recordar que una arista "saliente" representa la conexión de una palabra con su sucesora.
- Mayor *out-degree*: el grado mayor presente en la red del nodo considerando sus aristas "salientes".
- Promedio *total degree*: promedio de grados de todos los nodos de una red.
- Mayor *total degree*: el grado mayor presente en la red del nodo considerando todas sus aristas ("entrantes" y "salientes").

*Notar que todas las medidas anteriores (salvo la cantidad de aristas) consideran para su cálculo el peso de las aristas.*

```python
subgraphs_description = dict()
subgraphs = dict()

for language, graph in language_graphs.items():
    nodes_strongly_connected_component = max(nx.strongly_connected_components(language_graphs[language]), key = len)
    subgraph = language_graphs[language].subgraph(nodes_strongly_connected_component)
    subgraphs[language] = subgraph

    subgraphs_description[language] = {
        'Cantidad de nodos': subgraph.number_of_nodes(),
        'Cantidad de aristas': subgraph.number_of_edges(),
        'Peso total de las aristas': int(subgraph.size(weight = 'weight')),
        'Peso promedio de las aristas': subgraph.size(weight = 'weight') / subgraph.number_of_edges(),
        'Promedio de largo de caminos más cortos': nx.average_shortest_path_length(subgraph, weight = 'weight'),
        'Promedio in-degree': sum([degree for node, degree in subgraph.in_degree(weight = 'weight')]) / len(subgraph.in_degree(weight = 'weight')),
        'Mayor in-degree': max([degree for node, degree in subgraph.in_degree(weight = 'weight')]),
        'Promedio out-degree': sum([degree for node, degree in subgraph.out_degree(weight = 'weight')]) / len(subgraph.out_degree(weight = 'weight')),
        'Mayor out-degree': max([degree for node, degree in subgraph.out_degree(weight = 'weight')]),
        'Promedio total degree': sum([degree for node, degree in subgraph.degree(weight = 'weight')]) / len(subgraph.degree(weight = 'weight')),
        'Mayor total degree': max([degree for node, degree in subgraph.degree(weight = 'weight')])
    }

subgraphs_description_df = pd.DataFrame.from_dict(subgraphs_description, orient = 'columns')

p = pd.Series([0, 0, 0, 5, 5, 5, 0, 5, 0, 5, 0], subgraphs_description_df.index)
subgraphs_description_df.apply(lambda x: x.map(p.map(lambda x: '{{:0.{}f}}'.format(x).format).loc[x.name]), 1)
```

En primer lugar, se observa que las subredes generadas a partir de la componente conexa más grande, agrupa la gran mayoría de los nodos de la redes bases. Así por ejemplo, la red árabe que originalmente tiene 732 nodos, su componente conexa más grande está compuesta por 636 nodos.

A continuación, se muestra una tabla con el porcentaje de nodos que contiene cada subred conexa respecto de su red base. Se puede observar que la subred del español abarca el porcentaje más alto de nodos en comparación con el resto. En contraste, el chino es el idioma cuya subred si bien está compuesta por la mayoría de los nodos, es la que menos nodos porcentualmente contiene.

Al ser todos porcentajes altos, los resultados de las medidas aplicadas a las subredes son muy representativos de sus correspondientes redes bases.

```python
nodes_percentage_description = dict()

for language, subgraph in subgraphs.items():    
    nodes_percentage_description[language] = {
        'Porcentaje': (subgraph.number_of_nodes() / language_graphs[language].number_of_nodes()) * 100
    }

nodes_percentage_description_df = pd.DataFrame.from_dict(nodes_percentage_description, orient = 'index')
nodes_percentage_description_df.sort_values(['Porcentaje'], ascending = False)
```
