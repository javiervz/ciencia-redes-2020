El segundo problema que surge es que no todos estos idiomas tienen una estructura romance tradicional, es decir palabra-espacio-palabra, como es el caso del chino.

Los primeros idiomas que analizaremos serán los más comunes en términos de estructura y luego revisaremos formas de tokenizar correctamente idiomas con diferentes estructuras mediante el uso de librerías.

```python
file_prefix = "./datos/udhr_"
languages = {
    'spa': file_prefix + 'spa.txt',
    'eng': file_prefix + 'eng.txt',
    'fra': file_prefix + 'fra.txt',
    'rus': file_prefix + 'rus.txt',
    #'cmn': file_prefix + 'cmn_hans.txt',
    'arb': file_prefix + 'arb.txt'
}
```

```python
import re
import jieba # para idioma chino
#Abrimos un archivo tomando en cuenta el uso de utf8 para su parseo
corpuses = []
graphs = []

for lang in languages.values():
    with open(lang, encoding="utf8") as f:
        corpus = f.readlines()

    #las primeras 6 lineas corresponden a información del texto que no necesitamos para el análisis
    corpus = corpus[6:]
    #Eliminamos todos los elementos de saltos de líneas y las tabulaciones
    corpus = [x.strip() for x in corpus if x != '\n']
    corpus = [x for x in corpus if x != '']
    if lang in ['spa','eng','fra']:
        corpus = [x.lower() for x in corpus]


    #unimos todo en un gran string
    words = " ".join(corpus)
    #Eliminamos todos los símbolos y valores numéricos de cada línea
    punctuation = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    words = [letter for letter in words if letter not in punctuation]
    #words = [letter for letter in words if letter not in '1234567890']
    words = "".join(words)
    corpuses.append(words)
    
#Trabajamos con el idioma chino mandarin
# Para esto usamos la librería jieba que va a separar las oraciones en palabras separadas
# para esta separación estamos lamentablemente ligados a que tan bien funcione la librería
with open(file_prefix + 'cmn_hans.txt', encoding="utf8") as f:
    corpus = f.readlines()

    #las primeras 6 lineas corresponden a información del texto que no necesitamos para el análisis
    corpus = corpus[6:]
    #Eliminamos todos los elementos de saltos de líneas y las tabulaciones
    corpus = [x.strip() for x in corpus if x != '\n']
    corpus = [x for x in corpus if x != '']


    #unimos todo en un gran string
    words = " ".join(corpus)
    words = jieba.lcut(words)
    words = " ".join(words)
    #Eliminamos todos los símbolos y valores numéricos de cada línea
    punctuation = u'''!()-[]{};:'"\,<>./?@#$%^&*_~。》《，'''
    words = [letter for letter in words if letter not in punctuation]
    #words = [letter for letter in words if letter not in '1234567890']
    words = "".join(words)
    corpuses.append(words)
```

```python
import networkx as nx
from collections import Counter
edges_info = []
for words in corpuses:
    #Ahora creamos el grafo dirigido, para esto tomamos cada par de palabras y lo llamaremos una arísta
    words = words.split()
    edges = [(words[i].strip(), words[i+1].strip()) for i,x in enumerate(words) if i+1 < len(words)]
    '''así se ve un sample de 10 aristas
    [('Declaración', 'Universal'),
     ('Universal', 'de'),
     ('de', 'Derechos'),
     ('Derechos', 'Humanos'),
     ('Humanos', 'Adoptada'),
     ('Adoptada', 'y'),
     ('y', 'proclamada'),
     ('proclamada', 'por'),
     ('por', 'la'),
     ('la', 'Asamblea')]'''

    DG = nx.DiGraph()
    keys = list(Counter(edges).keys())
    values = list(Counter(edges).values())

    edges = [(x[0],x[1],float(values[i])) for i,x in enumerate(keys)]
    edges_info.append(edges)
    DG.add_weighted_edges_from(edges)
    graphs.append(DG)
```

```python
from operator import itemgetter
#hacemos los analisis realizados en el paper
## mayor componente conexa
LSN = []
LSE = []
TNEw = []
avgID = []
avgOD = []
bstID = []
bstOD = []
bgED = []
avgED = []

for i, DG in enumerate(graphs):
    NDG = nx.Graph(DG)
    largest_cc = max(nx.connected_components(NDG), key=len)
    LSN.append(len(largest_cc))
    LSG = DG.subgraph(largest_cc)
    LSE.append(len(list(LSG.edges())))
    TNEw.append(sum([int(x[2]) for x in edges_info[i]]))
    outD = [x[1] for x in list(DG.out_degree)]
    inD = [x[1] for x in list(DG.in_degree)]
    avgOD.append(sum(outD)/len(outD))
    avgID.append(sum(inD)/len(inD))
    bstID.append(max(inD))
    bstOD.append(max(outD))
    total_edges = [x[1] for x in list(DG.degree())]
    bgED.append(max(total_edges))
    avgED.append(sum(total_edges)/len(total_edges))
```

```python
from tabulate import tabulate

labels = [
"Nodes of LS",
"Edges of LS",
"Total number of edge weight in LS",
"Average in-degree",
"Biggest in-degree",
"Average out-degree",
"Biggest out-degree",
"Average total degree",
"Biggest total degree",
]

info = [
LSN,
LSE,
TNEw,
avgID,
bstID,
avgOD,
bstOD,
bgED,
avgED
]


data = []
for i in range(len(labels)):
    data.append([labels[i]] + info[i])

header = ["", "Español", "Inglés", "Francés", "Ruso", "Árabe", "Chino Mandarín"]
print(tabulate(data, headers=header, tablefmt="grid"))
```
